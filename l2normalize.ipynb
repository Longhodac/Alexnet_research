```python
import numpy as np
```

```python


def l2_normalize(vector):
    norm = np.sqrt(np.sum(vector**2))

    if norm == 0:
        return vector

    return vector / norm
```

```python
v = np.array([3, 4, 0])
normalized_v = l2_normalize(v)

normalized_v
```

```python
from sklearn.preprocessing import normalize 

v2 = normalize(v, norm='l2', axis=1)
v2
```

```python
high_var_array = np.array([1, 100, 1000, 10000])
low_var_array = np.array([2, 4, 6, 8])
```

```python
np.var(high_var_array), np.var(low_var_array)
```

```python
np.std(high_var_array), np.std(low_var_array)
```

```python
np.mean(high_var_array), np.mean(low_var_array)
```
